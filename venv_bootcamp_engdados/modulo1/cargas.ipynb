Com base no esquema abaixo precisamos converter nossos dados para o tipo apropriado então precisamos realizar as seguintes mudanças



**Tablea clientes**: Converter dados de `CPF`,`TelefoneFixo`,`TelefoneCelular` e `Número` para `string`. Além de no atributo `Sexo` converter `Masculino` para `M` e `Feminino` para `F`. 
Para garantir que nossos dados estão consistentes também vamos remover os espaços em branco de `Nome` e `Rua` e subsituir por `-`.

**Tablea apólices**: Converter dados de `DataInicio` e `Datafim` para `datetime.date`, `ValorFranquia` e `ValorCobertura` para `float`.  

**Tabela sinistros**: Converter dados `DataSinistro` para `datetime.date` e remover os espaços em branco de `Condutor` para substituir por `-`.
Aqui tambem, vamos precisar acrescentrar dados da tabela de apólices para sabermos qual é o apólice ativa de acordo com a `DataSinistro` e `CodCarro`.

**Tabela Carros**: Converter dados de `Ano` para `datetime.date` e remover os espaços em branco de `Placa` para substituir por "".

```python
#criando funções de preprocessamento
def change_type(df:pd.DataFrame,params:Dict):
    for column,new_type in params.items():
        df.loc[:,column] = df.loc[:,column].astype(new_type)
    return df

def convert_to_date(col:pd.Series,fmt:str):
    col = col.apply(lambda data: datetime.datetime.strptime(data,fmt).date())
    return col

def convert_to_datetime(col:pd.Series,fmt:str):
    col = col.apply(lambda data: datetime.datetime.strptime(data,fmt))
    return col
```

```python
# limpando colunas

# tabela clientes 
params_cliente = {
    "CPF":"str",
    "TelefoneFixo":"str",
    "TelefoneCelular":"str",
    "Numero":"str"
}

df_clientes = change_type(df_clientes,params_cliente)
df_clientes.loc[:,["Nome","Rua"]] = df_clientes.loc[:,["Nome","Rua"]].apply(lambda data:data.str.strip())
df_clientes.loc[:,["Nome","Rua"]] = df_clientes.loc[:,["Nome","Rua"]].apply(lambda data:data.str.replace(" ","-"))
df_clientes.loc[:,["Sexo"]] = df_clientes.loc[:,["Sexo"]].apply(lambda data:data.str.replace("Feminino","F"))
df_clientes.loc[:,["Sexo"]] = df_clientes.loc[:,["Sexo"]].apply(lambda data:data.str.replace("Masculino","M"))


#tabela apólices 
params_apolices = {
    "ValorFranquia":"float32",
    "ValorCobertura":"float32"
}

df_apolices = change_type(df_apolices,params_apolices)
df_apolices.loc[:,["DataInicio"]] = convert_to_date(df_apolices["DataInicio"],fmt="%Y-%m-%d")
df_apolices.loc[:,["Datafim"]] = convert_to_date(df_apolices["Datafim"],fmt="%Y-%m-%d")


df_sinistros.loc[:,["HoraSinistro"]] = df_sinistros["DataSinistro"] + " " +  df_sinistros["HoraSinistro"].astype("str")
df_sinistros.loc[:,["DataSinistro"]] = convert_to_date(df_sinistros["DataSinistro"],fmt="%Y-%m-%d")
df_sinistros.loc[:,["HoraSinistro"]] = convert_to_datetime(df_sinistros["HoraSinistro"],fmt="%Y-%m-%d %H")
df_sinistros.loc[:,["Condutor"]] = df_sinistros.loc[:,["Condutor"]].apply(lambda data:data.str.replace(" ","-"))

#tabela ano
df_carros.loc[:,["Ano"]] = df_carros.loc[:,"Ano"].astype("str")
df_carros.loc[:,["Ano"]] = convert_to_date(df_carros["Ano"].astype("str"),fmt="%Y")
df_carros.loc[:,["Placa"]] = df_carros.loc[:,["Placa"]].apply(lambda data:data.str.replace(" ",""))
```

```python
filter_apolices = df_apolices.filter(items=["CodApolice","Carro_CodCarro","DataInicio",'Datafim'])
df_sinistros["CodApoliceAtiva"]= np.nan

for item in df_sinistros.itertuples():
   dataSinistro = item.DataSinistro
   codCarro = item.Carro_CodCarro

   search_items = df_apolices.loc[df_apolices["Carro_CodCarro"]==codCarro,:]
   apolice_ativa = search_items[(search_items.DataInicio<dataSinistro)&(search_items.Datafim>dataSinistro)]["CodApolice"].values

   if len(apolice_ativa)!=0:
     df_sinistros.loc[df_sinistros["Carro_CodCarro"]==codCarro,"CodApoliceAtiva"]=apolice_ativa[0]
```

# 3 - Inserindo dados no banco de dados

## Inserindo dados na tabela cliente

```python
for cliente in df_clientes.itertuples():
        try:
            codCliente = cliente.CodCliente
            Nome = cliente.Nome
            CPF = cliente.CPF
            Sexo = cliente.Sexo
            Estado = cliente.Estado
            Cidade = cliente.Cidade
            Bairro = cliente.Bairro
            Rua = cliente.Rua
            Numero = cliente.Numero
            TelefoneFixo = cliente.TelefoneFixo
            TelefoneCelular = cliente.TelefoneCelular

            query = f'''INSERT INTO clientes (codClientes,Nome,CPF,Sexo,Estado,Cidade,Bairro,Rua,Numero,TelefoneFixo,TelefoneCelular) values ({codCliente},"{Nome}","{CPF}","{Sexo}",\
                "{Estado}","{Cidade}","{Bairro}","{Rua}","{Numero}","{TelefoneFixo}","{TelefoneCelular}");'''
            conn.execute(sqlalchemy.text(query))
            conn.commit()
            print(f"Dado de {Nome} inserido com sucesso!")
        
        except Exception as e:
                print(f"Não foi possível inserir o regristro {cliente.Nome}. O erro foi encontrado foi: {e}")
```

## Inserindo dados da tabela carros

```python
for carro in df_carros.itertuples():
        try:
            CodCarro = carro.ID
            Placa = carro.Placa
            Chassi  = carro.Chassi
            Marca = carro.Marca
            Modelo  = carro.Modelo
            Cor = carro.Cor
            Ano = carro.Ano
           
            query = f'''INSERT INTO carros (codCarros,Placa,Chassi,Marca,Modelo,Cor,Ano) values ( \
                "{CodCarro}","{Placa}","{Chassi}","{Marca}","{Modelo}","{Cor}","{Ano}");'''
            conn.execute(sqlalchemy.text(query))
            conn.commit()
            print(f"Dado de {CodCarro} inserido com sucesso!")
        
        except Exception as e:
                print(f"Não foi possível inserir o regristro {CodCarro}. O erro foi encontrado foi: {e}")
```

## Inserindo dados na tabela apólice

```python
# Inserindo dados da tabela apólices

for apolice in df_apolices.itertuples():
        try:
            CodApolice = apolice.CodApolice
            ValorCobertura = apolice.ValorCobertura
            ValorFranquia  = apolice.ValorFranquia
            DataInicio = apolice.DataInicio
            Datafim  = apolice.Datafim
            CodCliente = apolice.Cliente_CodCliente
            CodCarro = apolice.Carro_CodCarro
            query = f'''INSERT INTO apólices (codApólices,dataInicioVigência,dataFimVigência,valorTotalAssegurado,valorFranquia,codClientes,codCarros) values ( \
                "{CodApolice}","{DataInicio}","{Datafim}","{ValorCobertura}","{ValorFranquia}","{CodCliente}","{CodCarro}");'''
            conn.execute(sqlalchemy.text(query))
            conn.commit()
            print(f"Dado de {CodApolice} inserido com sucesso!")
        
        except Exception as e:
                print(f"Não foi possível inserir o regristro {CodApolice}. O erro foi encontrado foi: {e}")
```
